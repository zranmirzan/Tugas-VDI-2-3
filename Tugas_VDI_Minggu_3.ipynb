Tentukan data yang akan anda gunakan dalam visualisasi:
- Data yang saya gunakan didapat dari github(https://github.com/MainakRepositor/Datasets/blob/master/Cryptocurrency/bitcoin.csv)
- Dataset ini mencakup data dari cryptocurrency yaitu Bitcoin dari Tahun 2013-2020

Kolom dari dataset bitcoin mencakup hal ini:

- Date: hari dalam data.
- Open: buka harga untuk bitcoin.
- High: harga bitcoin tertinggi pada hari itu.
- Low: harga terendah bitcoin pada hari itu.
- Close: harga penutupan bitcoin.
- Volume: total volume perdagangan pada hari itu.
- Market Cap: total kapitalisasi pasar pada hari itu.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

bitcoin_data = pd.read_csv('/content/bitcoin.csv')
bitcoin_data
```

# 2

Tujuan
- Untuk memvisualisasikan tren harga Bitcoin dan metrik pasar dari waktu ke waktu.
- Aspek utama yang perlu dijelajahi adalah:

- Bagaimana harga Bitcoin (Open, High, Low, Close) berfluktuasi seiring waktu dalam hal ini rentang waktu 2013-2020.
- Hubungan antara volume perdagangan dan kapitalisasi pasar dengan harga Bitcoin.



User:
- Pedagang kripto (Trader): Tertarik untuk memahami pergerakan harga historis.
- Analis keuangan: Menjelajahi tren dan hubungan antara volume perdagangan dan kapitalisasi pasar.
- Masyarakat umum: Memahami bagaimana kinerja Bitcoin dari waktu ke waktu.

Fungsi:
- Analisis tren: Tunjukkan evolusi harga Bitcoin dari waktu ke waktu.
- Volume vs. Kapitalisasi Pasar: Visualisasikan bagaimana volume dan kapitalisasi pasar saling terkait dan bagaimana keduanya berubah seiring pergerakan harga.
- Menyoroti puncak dan penurunan: Tunjukkan momen yang penting dari perubahan harga yang signifikan dari 2013-2020.

Tone:
- Visualisasi harus menyajikan data yang jelas dan faktual dengan fokus pada presisi, melayani khalayak teknis dan umum.

# 3

```python
bitcoin_data['Date'] = pd.to_datetime(bitcoin_data['Date'])

bitcoin_data.set_index('Date', inplace=True)

plt.figure(figsize=(14, 8))

plt.plot(bitcoin_data.index, bitcoin_data['Close'], label='Close Price', color='blue')

max_close = bitcoin_data['Close'].max()
min_close = bitcoin_data['Close'].min()
max_close_date = bitcoin_data['Close'].idxmax()
min_close_date = bitcoin_data['Close'].idxmin()

plt.scatter(max_close_date, max_close, color='green', label=f'Highest Close: {max_close}', zorder=5)
plt.scatter(min_close_date, min_close, color='red', label=f'Lowest Close: {min_close}', zorder=5)

# Tambahkan title
plt.title('Bitcoin Close Price Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Close Price (USD)', fontsize=12)
plt.legend()

# Plot
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap to understand relationships between features
plt.figure(figsize=(8, 6))
corr = bitcoin_data[['Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Bitcoin Metrics', fontsize=14)
plt.show()
```
